```python
p = polyGraph("PL002.17", ph_df, 5)
```

```python
import numpy as np
```

```python
def solveForY(val, p):
    return((p - val).roots()[3])
```

```python
for col in rest_df.columns:
    for i in rest_df[col].dropna().values:
        print(np.absolute(solveForY(i, p)))
```

```python
def calcRest(rest_df, p):
    for col in rest_df.columns:
        out_list = []
        for i in rest_df[col].dropna().values:
            out_list.append(np.absolute(solveForY(i, p)))
        filename = str(col) + ".txt"
        f = open(filename, 'w')
        for item in out_list:
            f.write(f"{item}\n")  
```

```python
calcRest(rest_df, p)
```
